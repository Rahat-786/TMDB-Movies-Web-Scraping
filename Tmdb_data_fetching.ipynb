```python
import requests
import pandas as pd
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=ac48a6b2b92624250c515f0924cccd99&language=en-US&page=1").json()
```

```python
response
```

```python
popularity = []
vote_count = []
original_lanuage = []
title = []
vote_average = []
overview = []
release_date = []


for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_lanuage.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
    
# creating a dictionary

d = {'title' : title,
     'release_date' : release_date,
    'original_language' : original_lanuage,
    'popularity' : popularity,
    'overview' : overview,
    'vote_average' : vote_average,
    'vote_count' : vote_count,}

df = pd.DataFrame(d)
df

```

```python
final = pd.DataFrame()

for j in range(1, 501):
    response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=ac48a6b2b92624250c515f0924cccd99&language=en-US&page={}".format(j)).json()
    
    popularity = []
    vote_count = []
    original_lanuage = []
    title = []
    vote_average = []
    overview = []
    release_date = []


    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_lanuage.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])


    # creating a dictionary

    d = {'title' : title,
         'release_date' : release_date,
        'original_language' : original_lanuage,
        'popularity' : popularity,
        'overview' : overview,
        'vote_average' : vote_average,
        'vote_count' : vote_count,}

    df = pd.DataFrame(d)
    final = final.append(df, ignore_index=True)
    

```
